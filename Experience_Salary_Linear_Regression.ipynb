```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('Experience-Salary.csv')
df
```

```python
plt.xlabel("exp")
plt.ylabel("salary")
plt.scatter(df['exp(in months)'], df['salary(in thousands)'],color='red', marker='.')
```

new_df is likely a Series (1D).

But fit() needs X to be 2D ([n_samples, n_features]).

sal (target y) can be 1D.

so you can't simply say, new_df=df['exp(in months)']

```python
new_df = df.drop('salary(in thousands)', axis=1)
sal =df['salary(in thousands)']
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,sal)
```

```python
reg.predict([[12]])
```
