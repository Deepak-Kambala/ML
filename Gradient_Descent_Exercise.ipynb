```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def gradient_descent(x, y, rate=0.0001, iterations=1000):
    m = b = 0
    n = len(x)
    for _ in range(iterations):
        y_pred = m * x + b
        md = -(2/n) * np.sum(x * (y - y_pred))
        bd = -(2/n) * np.sum(y - y_pred)
        m = m - rate * md
        b = b - rate * bd
    return m, b
```

```python
df = pd.read_csv('test_scores.csv')
df
```

```python
math = df['math'].to_numpy()
cs = df['cs'].to_numpy()
```

```python
m_final, b_final = gradient_descent(math,cs)
```

```python
print(f"Slope (m): {m_final}, Intercept (b): {b_final}")
```

```python
plt.scatter(math, cs, color='red', marker='+', linewidth=5)
plt.plot(math, m_final*math + b_final, color='green')
plt.xlabel("Math scores")
plt.ylabel("CS scores")
plt.show()
```
